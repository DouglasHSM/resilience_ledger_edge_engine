````python
%%writefile database.py
import aiosqlite, asyncio, json, os
from datetime import datetime

DB_PATH = "data/ledger.db"

async def init_db():
    """
    [SYS] INITIALIZE LEDGER
    Creates the main inventory table and the immutable audit log table.
    Ensures the data directory exists before deployment.
    """
    os.makedirs(os.path.dirname(DB_PATH), exist_ok=True)
    async with aiosqlite.connect(DB_PATH) as db:
        # Main Tactical Inventory
        await db.execute('''CREATE TABLE IF NOT EXISTS inventory_ledger (
            id INTEGER PRIMARY KEY AUTOINCREMENT, 
            item TEXT, 
            category TEXT, 
            qty INTEGER, 
            unit TEXT, 
            exp_date TEXT, 
            status TEXT, 
            operator TEXT, 
            timestamp TEXT)''')
            
        # Immutable Audit Log for accountability (Tracing HIL validations and overrides)
        await db.execute('''CREATE TABLE IF NOT EXISTS audit_log (
            id INTEGER PRIMARY KEY AUTOINCREMENT, 
            timestamp TEXT NOT NULL, 
            badge TEXT NOT NULL,
            action TEXT NOT NULL, 
            ai_raw TEXT, 
            final_value TEXT, 
            confidence INTEGER, 
            diff TEXT)''')
        await db.commit()

async def insert_inventory_item(data, badge):
    """
    [SYS] INSERT RECORD
    Saves a new supply entry to the database and logs the 'APPROVE' action.
    Triggered after the operator validates the AI's visual extraction.
    """
    timestamp = datetime.now().isoformat()
    async with aiosqlite.connect(DB_PATH) as db:
        await db.execute("INSERT INTO inventory_ledger (item, category, qty, unit, exp_date, status, operator, timestamp) VALUES (?,?,?,?,?,?,?,?)",
            (data['item'], data.get('category', 'General'), data['qty'], data['unit'], data.get('exp_date', 'N/A'), '● OK', badge, timestamp))
            
        await db.execute("INSERT INTO audit_log (timestamp, badge, action, ai_raw, final_value, confidence, diff) VALUES (?,?,?,?,?,?,?)",
            (timestamp, badge, "APPROVE", data.get('ai_raw', 'N/A'), json.dumps(data), data.get('confidence', 0), "ORIGINAL ENTRY"))
        await db.commit()

async def update_inventory_item(item_id, data):
    """
    [SYS] UPDATE RECORD
    Modifies an existing record. Captures the exact difference (old qty vs new qty)
    in the audit log to prevent untraceable tampering.
    """
    timestamp = datetime.now().isoformat()
    badge = data.get('operator', 'SYS_GHOST')
    async with aiosqlite.connect(DB_PATH) as db:
        cursor = await db.execute("SELECT item, qty FROM inventory_ledger WHERE id=?", (item_id,))
        old = await cursor.fetchone()
        
        await db.execute("UPDATE inventory_ledger SET item=?, category=?, qty=?, unit=?, exp_date=?, operator=?, timestamp=? WHERE id=?",
            (data['item'], data.get('category', 'General'), data['qty'], data['unit'], data.get('exp_date', 'N/A'), badge, timestamp, item_id))
            
        diff = f"EDITED: {old[0]}({old[1]}) -> {data['item']}({data['qty']})" if old else "MANUAL EDIT"
        
        await db.execute("INSERT INTO audit_log (timestamp, badge, action, ai_raw, final_value, confidence, diff) VALUES (?,?,?,?,?,?,?)",
            (timestamp, badge, "EDIT", "HUMAN_OVERRIDE", json.dumps(data), 100, diff))
        await db.commit()

async def delete_inventory_item(item_id, badge="SYS_GHOST"):
    """
    [SYS] PURGE RECORD
    Removes the item from the active inventory interface, but immortalizes 
    the deletion event in the audit log.
    """
    timestamp = datetime.now().isoformat()
    async with aiosqlite.connect(DB_PATH) as db:
        cursor = await db.execute("SELECT item, qty FROM inventory_ledger WHERE id=?", (item_id,))
        old = await cursor.fetchone()
        
        await db.execute("DELETE FROM inventory_ledger WHERE id=?", (item_id,))
        
        diff = f"PURGED: {old[0]} ({old[1]} units)" if old else "PURGED"
        
        await db.execute("INSERT INTO audit_log (timestamp, badge, action, ai_raw, final_value, confidence, diff) VALUES (?,?,?,?,?,?,?)",
            (timestamp, badge, "DELETE", "N/A", "REMOVED", 100, diff))
        await db.commit()

async def get_all_inventory():
    """
    [SYS] FETCH LEDGER
    Retrieves the entire active inventory, ordered by the most recent additions.
    """
    async with aiosqlite.connect(DB_PATH) as db:
        db.row_factory = aiosqlite.Row
        cursor = await db.execute("SELECT * FROM inventory_ledger ORDER BY id DESC")
        return [dict(row) for row in await cursor.fetchall()]
````

````python
import sqlite3
import os
from datetime import datetime

DB_PATH = "data/ledger.db"

def get_db_connection():
    """Establishes a secure connection to the active Edge ledger."""
    return sqlite3.connect(DB_PATH)

def auto_categorize(item_name: str) -> str:
    """Tactical helper to automatically categorize items based on keywords."""
    name_lower = item_name.lower()
    if any(k in name_lower for k in ['amoxicillin', 'paracetamol', 'pill', 'med', 'fluid']):
        return 'Med'
    elif any(k in name_lower for k in ['gauze', 'suture', 'scalpel', 'tourniquet', 'kit']):
        return 'Surgical'
    elif any(k in name_lower for k in ['ration', 'water', 'food', 'mre']):
        return 'Ration'
    elif any(k in name_lower for k in ['mask', 'glove', 'suit', 'ppe']):
        return 'PPE'
    return 'General'

print("=========================================================")
print("● RESILIENCE LEDGER // CORE TERMINAL OPERATIONAL NODE")
print("  DIRECT CRUD UPLINK SYSTEM // AIR-GAPPED DISASTER MODE")
print("=========================================================")
print("Available Commands:")
print("  /add [qty] [unit] [item_name]")
print("  /edit [ID] [new_qty]")
print("  /del [ID]")
print("  /list")
print("  /exit")
print("=========================================================")

while True:
    try:
        # Command line prompt styled for a tactical operator
        user_input = input("\n[operator@resilience-edge] $ ").strip()
        
        if not user_input:
            continue
            
        if user_input.lower() == "/exit":
            print("[!] Terminating secure core uplink session...")
            break
            
        # 🟢 1. CREATE COMMAND: /add [qty] [unit] [item_name]
        if user_input.startswith("/add "):
            try:
                parts = user_input[5:].split(" ", 2)
                if len(parts) < 3:
                    print("❌ SYNTAX ERROR. Expected: /add [qty] [unit] [item_name]")
                    continue
                    
                qty = int(parts[0])
                unit = parts[1]
                item_name = parts[2]
                category = auto_categorize(item_name)
                timestamp = datetime.now().isoformat()
                
                conn = get_db_connection()
                cursor = conn.cursor()
                cursor.execute("""
                    INSERT INTO inventory_ledger (item, category, qty, unit, exp_date, status, operator, timestamp)
                    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
                """, (item_name, category, qty, unit, "2028-12-01", "● OK", "CORE_TERMINAL", timestamp))
                conn.commit()
                conn.close()
                
                print(f"✅ SUCCESS: Automatically categorized as '{category}' and injected into SQLite database.")
            except ValueError:
                print("❌ VALUATION ERROR: Quantity must be a valid integer.")
                
        # 🟡 2. UPDATE COMMAND: /edit [ID] [new_qty]
        elif user_input.startswith("/edit "):
            try:
                parts = user_input[6:].split(" ")
                if len(parts) < 2:
                    print("❌ SYNTAX ERROR. Expected: /edit [ID] [new_qty]")
                    continue
                    
                record_id = int(parts[0])
                new_qty = int(parts[1])
                
                conn = get_db_connection()
                cursor = conn.cursor()
                
                # Check if item exists before editing
                cursor.execute("SELECT item FROM inventory_ledger WHERE id = ?", (record_id,))
                item_exists = cursor.fetchone()
                
                if not item_exists:
                    print(f"❌ NOT FOUND: Record ID {record_id} does not exist in the active ledger.")
                    conn.close()
                    continue
                    
                cursor.execute("UPDATE inventory_ledger SET qty = ? WHERE id = ?", (new_qty, record_id))
                conn.commit()
                conn.close()
                
                print(f"⚠️ AUDIT TRACE TRIGGERED: Updated volume for Item ID {record_id} to {new_qty} units.")
            except ValueError:
                print("❌ VALUATION ERROR: ID and Quantity must be valid integers.")

        # 🔴 3. DELETE COMMAND: /del [ID]
        elif user_input.startswith("/del "):
            try:
                record_id = int(user_input[5:].strip())
                
                conn = get_db_connection()
                cursor = conn.cursor()
                
                # Check existence for accountability
                cursor.execute("SELECT item FROM inventory_ledger WHERE id = ?", (record_id,))
                item_exists = cursor.fetchone()
                
                if not item_exists:
                    print(f"❌ NOT FOUND: Record ID {record_id} does not exist.")
                    conn.close()
                    continue
                    
                cursor.execute("DELETE FROM inventory_ledger WHERE id = ?", (record_id,))
                conn.commit()
                conn.close()
                
                print(f"🔥 PURGE LOGGED: Record ID {record_id} physically removed from active ledger index.")
            except ValueError:
                print("❌ VALUATION ERROR: Record ID must be a valid integer.")

        # 🔵 EXTRA COMMAND: /list (To verify live updates easily during video)
        elif user_input.strip() == "/list":
            conn = get_db_connection()
            cursor = conn.cursor()
            cursor.execute("SELECT id, item, category, qty, unit, status FROM inventory_ledger ORDER BY id DESC LIMIT 5")
            rows = cursor.fetchall()
            conn.close()
            
            print("\n--- ACTIVE INVENTORY LEDGER ROWS ---")
            if not rows:
                print("Ledger currently empty.")
            for r in rows:
                print(f"ID: {r[0]} | Item: {r[1]} | Cat: {r[2]} | Qty: {r[3]} {r[4]} | Status: {r[5]}")
            print("-------------------------------------")
            
        else:
            print("❌ COMMAND UNRECOGNIZED. Use /add, /edit, /del, or /list")
            
    except KeyboardInterrupt:
        print("\n[!] Session interrupted by operator. Safe exit.")
        break
````

````python
%%writefile static/app.js
'use strict';
/* ============================================================
   RESILIENCE LEDGER — Tactical Cortex v2.0
   Modular frontend for humanitarian aid inventory management.
   ============================================================ */
(() => {

// --- Config ---
// PROMPT MILITAR BLINDADO: Impede o loop "own own own" forçando o formato estrito
const LLM_PROMPT = 'TACTICAL DIRECTIVE: Identify the MEDICATION or SUPPLY in the image, ignoring background noise. Extract expiration date if visible. You MUST return IMMEDIATELY and ONLY the pure JSON. NO reasoning, NO intros, NO extra words. Exact format: {"item": "name", "qty": 1, "unit": "type", "exp_date": "YYYY-MM-DD", "confidence": 99}';
const CHAT_SYSTEM = '[TACTICAL DIRECTIVE] You are the Offline Tactical Cortex of the Resilience Ledger. Respond strategically to the operator.';

// --- Helpers ---
const $ = id => document.getElementById(id);
const $$ = sel => document.querySelectorAll(sel);
const now = () => new Date().toLocaleTimeString('en-US', {hour12:false});

// --- State ---
const state = {
  badge: null, startTime: null, currentScreen: 'screen-dashboard',
  previousScreen: 'screen-dashboard', inventory: [], videoStream: null,
  editId: null, hilData: null,
  stats: { processed:0, approved:0, rejected:0, edits:0, sumConf:0, confMin:100 },
  charts: { inv:null, depl:null, hourly:null }
};

// --- API Client ---
const api = {
  async get(url) {
    const r = await fetch(url); if (!r.ok) throw new Error(r.statusText);
    return r.json();
  },
  async post(url, body) {
    const r = await fetch(url, {method:'POST', headers:{'Content-Type':'application/json'}, body:JSON.stringify(body)});
    return r.json();
  },
  async put(url, body) {
    const r = await fetch(url, {method:'PUT', headers:{'Content-Type':'application/json'}, body:JSON.stringify(body)});
    return r.json();
  },
  async del(url) { return (await fetch(url, {method:'DELETE'})).json(); },

  loadInventory: async () => {
    try { const d = await api.get('/api/inventory'); if (d.inventory) { state.inventory = d.inventory; dashboard.refresh(); } } catch(e) { console.warn('API offline', e); }
  },
  checkHealth: async () => {
    try {
      const d = await api.get('/api/health');
      $('diag-llm').textContent = d.status === 'online' ? 'ONLINE' : 'OFFLINE';
      $('diag-llm').className = 'status-val ' + (d.status === 'online' ? 'online' : 'error');
      $('diag-vision').textContent = d.status === 'online' ? 'ONLINE' : 'OFFLINE';
      $('diag-vision').className = 'status-val ' + (d.status === 'online' ? 'online' : 'error');
    } catch(e) {
      $('diag-llm').textContent = 'OFFLINE'; $('diag-llm').className = 'status-val error';
      $('diag-vision').textContent = 'OFFLINE'; $('diag-vision').className = 'status-val error';
    }
  }
};

// --- Clock ---
function tick() {
  const t = $('diag-time'); if (t) t.textContent = new Date().toISOString().replace('T',' ').substring(0,19);
  if (state.startTime) {
    const ms = Date.now() - state.startTime;
    const h = Math.floor(ms/36e5), m = Math.floor((ms%36e5)/6e4), s = Math.floor((ms%6e4)/1e3);
    const dur = `${String(h).padStart(2,'0')}h${String(m).padStart(2,'0')}m`;
    if ($('shift-dur')) $('shift-dur').textContent = dur + String(s).padStart(2,'0') + 's';
  }
}

// --- Flash ---
function flash(msg, color) {
  const el = $('flash-msg'); if (!el) return;
  el.textContent = '● ' + (msg || 'OPERATION SUCCESSFUL');
  if (color) el.style.background = color;
  el.style.display = 'block'; setTimeout(() => el.style.display = 'none', 2200);
}

// --- Navigation ---
const nav = {
  goto(id) {
    if (state.currentScreen !== 'screen-hil') state.previousScreen = state.currentScreen;
    $$('.main-area .screen').forEach(s => s.classList.remove('active'));
    const target = $(id); if (target) target.classList.add('active');
    state.currentScreen = id;
    $$('.nav-item').forEach(n => {
      n.classList.toggle('active', n.dataset.screen === id);
    });
    if (id === 'screen-camera') camera.start(); else camera.stop();
    if (id === 'screen-terminal') { const h = $('terminal-history'); if (h) h.scrollTop = h.scrollHeight; }
  },
  init() {
    $$('.nav-item').forEach(n => {
      n.addEventListener('click', () => nav.goto(n.dataset.screen));
      n.addEventListener('keydown', e => { if (e.key === 'Enter') nav.goto(n.dataset.screen); });
    });
  }
};

// --- Login ---
function doLogin() {
  state.badge = $('login-operator').value || 'GUEST';
  state.startTime = Date.now();
  $('sb-operator').textContent = state.badge;
  if ($('shift-op')) $('shift-op').textContent = state.badge;
  if ($('shift-start')) $('shift-start').textContent = new Date().toLocaleTimeString('en-US', {hour12:false});
  $('screen-login').classList.remove('active');
  $('main-app').style.display = 'flex';
  nav.goto('screen-dashboard');
  api.loadInventory();
  charts.init();
  comms.add('System online. Operator ' + state.badge + ' authenticated.', 'system');
}

// --- Comms / Messaging ---
const comms = {
  add(text, type) {
    const t = now();
    const cls = type === 'system' ? 'msg-system' : 'msg-operator';
    const prefix = type === 'system' ? '[SYS]' : '[' + (state.badge||'OP') + ']';
    const html = `<div class="${cls}">${prefix} ${t} | ${text}</div>`;
    const f1 = $('comms-feed'), f2 = $('terminal-history');
    if (f1) { f1.insertAdjacentHTML('beforeend', html); $('comms-scroll').scrollTop = f1.scrollHeight; }
    if (f2) { f2.insertAdjacentHTML('beforeend', html); f2.scrollTop = f2.scrollHeight; }
  },
  addAI(text) {
    const t = now();
    const html = `<div class="msg-ai">[CORE] ${t} | ${text}</div>`;
    const f1 = $('comms-feed'), f2 = $('terminal-history');
    if (f1) { f1.insertAdjacentHTML('beforeend', html); $('comms-scroll').scrollTop = f1.scrollHeight; }
    if (f2) { f2.insertAdjacentHTML('beforeend', html); f2.scrollTop = f2.scrollHeight; }
  }
};

// --- Categorizer ---
function categorize(name) {
  const n = name.toLowerCase();
  if (/amoxicillin|dipyrone|paracetamol|naproxen|medication|pill|tablet|ibuprofen|antibiotic/.test(n)) return 'Med';
  if (/scalpel|glove|syringe|gauze|suture|surgical/.test(n)) return 'Surgical';
  if (/rice|water|food|mre|ration|beans|milk/.test(n)) return 'Ration';
  if (/mask|ppe|helmet|goggles|protection/.test(n)) return 'PPE';
  return 'General';
}

// --- LLM Response Parser ---
function parseLLM(text) {
  try {
    const m = text.match(/\{[\s\S]*?\}/);
    if (m) {
      const o = JSON.parse(m[0]);
      let c = parseFloat(o.confidence); if (c > 0 && c <= 1) c *= 100;
      return { item: o.item||'Unknown', qty: parseInt(o.qty)||1, unit: o.unit||'units', exp_date: o.exp_date||'N/A', confidence: Math.round(c)||50 };
    }
  } catch(e) {}
  return { item:'Extraction Failed', qty:1, unit:'units', exp_date:'N/A', confidence:20 };
}

// --- Camera Module ---
const camera = {
  async start() {
    if (state.videoStream) return;
    try {
      state.videoStream = await navigator.mediaDevices.getUserMedia({video:{facingMode:'environment'}});
      $('video-element').srcObject = state.videoStream;
    } catch(e) { comms.add('Camera access denied: ' + e.message, 'system'); }
  },
  stop() {
    if (state.videoStream) { state.videoStream.getTracks().forEach(t => t.stop()); state.videoStream = null; }
  },
  async capture() {
    const v = $('video-element'), c = $('preview-canvas');
    if (!v || !c || !state.videoStream) return;
    c.width = v.videoWidth; c.height = v.videoHeight;
    c.getContext('2d').drawImage(v, 0, 0);
    $('cam-processing').style.display = 'flex'; $('cam-progress').style.width = '30%';
    try {
      const b64 = c.toDataURL('image/jpeg', 0.8).split(',')[1];
      const payload = { messages:[{role:'user',content:[{type:'text',text:LLM_PROMPT},{type:'image_url',image_url:{url:'data:image/jpeg;base64,'+b64}}]}], max_tokens:150, temperature:0.1, stop:['</s>','<eos>','<|im_end|>'] };
      const data = await api.post('/api/chat', payload);
      $('cam-progress').style.width = '90%';
      const parsed = parseLLM(data.choices[0].message.content);
      hil.open({...parsed, ai_raw: data.choices[0].message.content});
    } catch(e) { comms.add('Visual extraction failed: ' + e.message, 'system'); }
    finally { $('cam-processing').style.display = 'none'; $('cam-progress').style.width = '0%'; }
  }
};

// --- HIL (Human-in-the-Loop) ---
const hil = {
  open(data) {
    state.hilData = data;
    state.stats.processed++; state.stats.sumConf += data.confidence;
    if (data.confidence < state.stats.confMin) state.stats.confMin = data.confidence;
    nav.goto('screen-hil');
    $('hil-item').value = data.item; $('hil-qty').value = data.qty;
    $('hil-unit').value = data.unit; $('hil-exp').value = data.exp_date;
    $('hil-raw').value = data.ai_raw || '';
    $('hil-conf-fill').style.width = data.confidence + '%';
    $('hil-conf-text').textContent = data.confidence + '%';
    const isOk = data.confidence >= 60 || state.editId;
    $('hil-conf-fill').className = 'confidence-fill ' + (isOk ? 'conf-high' : 'conf-low');
    $('btn-approve').disabled = !isOk;
    $('btn-approve').textContent = state.editId ? '[ENTER] SAVE CHANGES' : '[ENTER] CONFIRM REGISTRY';
    shift.update();
  },
  async approve() {
    const item = $('hil-item').value, qty = parseInt($('hil-qty').value),
          unit = $('hil-unit').value, exp = $('hil-exp').value;
    const payload = { item, qty, unit, exp_date: exp, category: categorize(item), operator: state.badge };
    try {
      if (state.editId) {
        await api.put('/api/inventory/' + state.editId, payload);
      } else {
        await api.post('/api/inventory', payload);
      }
      state.stats.approved++; shift.update();
      state.editId = null;
      flash(); comms.add('Item logged: ' + item + ' x' + qty, 'system');
      await api.loadInventory();
      nav.goto('screen-ledger');
    } catch(e) { comms.add('LEDGER ERROR: ' + e.message, 'system'); }
  },
  reject() { state.stats.rejected++; shift.update(); state.editId = null; nav.goto(state.previousScreen); }
};

// --- Dashboard ---
const dashboard = {
  refresh() {
    const inv = state.inventory;
    $('dash-items').textContent = inv.length;
    $('sb-items').textContent = inv.length;
    if (inv.length > 0) $('dash-latest').textContent = inv[0].item;
    const crit = inv.filter(i => i.qty < 10);
    $('dash-deficit').textContent = crit.length;
    $('sb-deficit').textContent = 'DEFICIT: ' + crit.length;
    // Table
    const tbody = document.querySelector('#inventory-table tbody');
    if (tbody) {
      tbody.innerHTML = '';
      inv.forEach(i => {
        const isCrit = i.qty < 10;
        const tr = document.createElement('tr');
        tr.innerHTML = `<td class="item-name">${i.item}</td><td>${i.category}</td>` +
          `<td class="${isCrit?'qty-critical':'qty-ok'}">${i.qty}</td><td>${i.unit}</td>` +
          `<td>${i.exp_date||'N/A'}</td><td class="${isCrit?'status-crit':'status-ok'}">${isCrit?'CRITICAL':'VERIFIED'}</td>` +
          `<td style="text-align:right"><button class="action-btn" data-edit="${i.id}">EDIT</button>` +
          `<button class="action-btn action-btn-del" data-del="${i.id}">PURGE</button></td>`;
        tbody.appendChild(tr);
      });
    }
    // Charts
    if (state.charts.inv) {
      const cats = ['Med','Surgical','Ration','PPE','General'];
      state.charts.inv.data.datasets[0].data = cats.map(c => inv.filter(i=>i.category===c).reduce((s,i)=>s+i.qty,0));
      state.charts.inv.update();
    }
    if (state.charts.depl) {
      const total = inv.reduce((s,i)=>s+i.qty,0);
      state.charts.depl.data.datasets[0].data = [total, Math.floor(total*.7), Math.floor(total*.3), Math.floor(total*.05)];
      state.charts.depl.update();
    }
    if (state.charts.hourly) {
      const total = inv.reduce((s,i)=>s+i.qty,0);
      state.charts.hourly.data.datasets[0].data = [Math.floor(total*.2), Math.floor(total*.4), Math.floor(total*.3), Math.floor(total*.1)];
      state.charts.hourly.update();
    }
  }
};

// --- Charts ---
const charts = {
  init() {
    if (typeof Chart === 'undefined') return;
    Chart.defaults.color = '#64748b';
    Chart.defaults.font.family = "'Courier New', monospace";
    const grid = {color:'rgba(255,255,255,0.05)'};
    const noGrid = {display:false};
    const opts = {responsive:true, maintainAspectRatio:false, plugins:{legend:{display:false}}};

    const c1 = $('chart-inventory')?.getContext('2d');
    if (c1) state.charts.inv = new Chart(c1, {type:'bar', data:{labels:['Med','Surgical','Ration','PPE','General'], datasets:[{data:[0,0,0,0,0], backgroundColor:'#00e5ff', borderRadius:4}]}, options:{...opts, indexAxis:'y', scales:{x:{grid, ticks:{precision:0}}, y:{grid:noGrid}}}});

    const c2 = $('chart-depletion')?.getContext('2d');
    if (c2) state.charts.depl = new Chart(c2, {type:'line', data:{labels:['T-0','T+1h','T+2h','T+3h'], datasets:[{label:'Vol', data:[0,0,0,0], borderColor:'#ef4444', backgroundColor:'rgba(239,68,68,0.1)', fill:true, tension:.3}]}, options:{...opts, scales:{y:{ticks:{precision:0}}}}});

    const c3 = $('chart-hourly')?.getContext('2d');
    if (c3) state.charts.hourly = new Chart(c3, {type:'bar', data:{labels:['Alpha','Bravo','Charlie','Current'], datasets:[{data:[0,0,0,0], backgroundColor:'#10b981', borderRadius:4}]}, options:{...opts, scales:{y:{ticks:{precision:0}}}}});
  }
};

// --- Shift Report ---
const shift = {
  update() {
    const s = state.stats, p = s.processed;
    if ($('shift-ops')) $('shift-ops').textContent = `${p} processed | ${s.approved} approved | ${s.rejected} rejected`;
    if ($('shift-edits')) $('shift-edits').textContent = `${s.edits} fields corrected`;
    const avg = p > 0 ? Math.round(s.sumConf/p) : 0;
    const min = p > 0 ? s.confMin : 0;
    if ($('shift-conf')) $('shift-conf').textContent = `avg ${avg}% | min ${min}%`;
  },
  close() { if (confirm('End shift and reload?')) location.reload(); }
};

// --- Terminal / Chat ---
const terminal = {
  async send() {
    const inp = $('terminal-input'), text = inp.value.trim();
    if (!text) return;
    inp.value = ''; inp.disabled = true;
    comms.add(text, 'operator');

    // Command parser
    if (text.startsWith('/add '))  return terminal.cmdAdd(text, inp);
    if (text.startsWith('/edit ')) return terminal.cmdEdit(text, inp);
    if (text.startsWith('/del '))  return terminal.cmdDel(text, inp);

    // LLM chat
    const msgId = 'msg-' + Date.now();
    $('terminal-history').insertAdjacentHTML('beforeend', `<div id="${msgId}" class="msg-loading">[SYS] CORTEX PROCESSING<span class="loading-dots"></span></div>`);
    $('terminal-history').scrollTop = $('terminal-history').scrollHeight;
    try {
      const payload = {messages:[{role:'user', content: CHAT_SYSTEM + '\n\n[OPERATOR]: ' + text}], max_tokens:300, temperature:0.2, stop:['</s>','<eos>','<|im_end|>']};
      const data = await api.post('/api/chat', payload);
      const el = $(msgId); if (el) { el.className = 'msg-ai'; el.textContent = '[CORE] ' + now() + ' | ' + data.choices[0].message.content; }
    } catch(e) { const el = $(msgId); if (el) { el.className = 'msg-ai'; el.textContent = '[SYS] CORTEX UPLINK FAILURE.'; } }
    finally { inp.disabled = false; inp.focus(); }
  },
  async cmdAdd(text, inp) {
    const p = text.split(' ');
    if (p.length >= 4) {
      const name = p.slice(3).join(' ');
      await api.post('/api/inventory', {item:name, qty:parseInt(p[1]), unit:p[2], exp_date:'N/A', category:categorize(name), operator:state.badge});
      comms.add(`CMD: ${p[1]} ${p[2]} of ${name} inserted.`, 'system');
      await api.loadInventory(); nav.goto('screen-ledger');
    } else { comms.add('Syntax: /add [qty] [unit] [name]', 'system'); }
    inp.disabled = false; inp.focus();
  },
  async cmdEdit(text, inp) {
    const p = text.split(' '), id = parseInt(p[1]), qty = parseInt(p[2]);
    const item = state.inventory.find(i => i.id === id);
    if (item && !isNaN(qty)) {
      await api.put('/api/inventory/'+id, {item:item.item, qty, unit:item.unit, exp_date:item.exp_date, category:item.category, operator:state.badge});
      comms.add('CMD: Record ID '+id+' updated to qty='+qty, 'system');
      await api.loadInventory(); nav.goto('screen-ledger');
    } else { comms.add('Syntax: /edit [ID] [NEW_QTY]', 'system'); }
    inp.disabled = false; inp.focus();
  },
  async cmdDel(text, inp) {
    const id = parseInt(text.split(' ')[1]);
    if (id) {
      await api.del('/api/inventory/'+id+'?badge='+(state.badge||'SYSTEM'));
      comms.add('CMD: Record ID '+id+' purged.', 'system');
      await api.loadInventory(); nav.goto('screen-ledger');
    } else { comms.add('Syntax: /del [ID]', 'system'); }
    inp.disabled = false; inp.focus();
  },
  async handleUpload(file) {
    comms.add('Analyzing file: ' + file.name, 'operator');
    const msgId = 'msg-' + Date.now();
    $('terminal-history').insertAdjacentHTML('beforeend', `<div id="${msgId}" class="msg-loading">[SYS] EXTRACTING VISUAL DATA<span class="loading-dots"></span></div>`);
    try {
      const b64 = await new Promise(r => { const fr = new FileReader(); fr.onload = e => r(e.target.result.split(',')[1]); fr.readAsDataURL(file); });
      const payload = {messages:[{role:'user',content:[{type:'text',text:LLM_PROMPT},{type:'image_url',image_url:{url:'data:image/jpeg;base64,'+b64}}]}], max_tokens:150, temperature:0.1, stop:['</s>','<eos>','<|im_end|>']};
      const data = await api.post('/api/chat', payload);
      const el = $(msgId); if (el) el.remove();
      const parsed = parseLLM(data.choices[0].message.content);
      hil.open({...parsed, ai_raw: data.choices[0].message.content});
    } catch(e) { const el = $(msgId); if (el) { el.className = 'msg-ai'; el.textContent = '[SYS] ERROR PROCESSING VISUAL.'; } }
    $('file-input-chat').value = '';
  }
};

// --- Export ---
function downloadCSV() {
  let csv = 'ID,ITEM,CATEGORY,QTY,UNIT,EXP_DATE,STATUS,OPERATOR\n';
  state.inventory.forEach(i => csv += `${i.id},${i.item},${i.category},${i.qty},${i.unit},${i.exp_date},${i.status},${i.operator}\n`);
  const a = document.createElement('a');
  a.href = URL.createObjectURL(new Blob([csv], {type:'text/csv'}));
  a.download = `ledger_export_${state.badge||'SYS'}_${Date.now()}.csv`; a.click();
}

// --- Keyboard Shortcuts ---
function handleKeys(e) {
  if (['INPUT','TEXTAREA'].includes(document.activeElement.tagName)) {
    if (e.key === 'Enter' && document.activeElement.id === 'terminal-input') terminal.send();
    return;
  }
  const map = {'1':'screen-dashboard','2':'screen-camera','3':'screen-ledger','4':'screen-analysis','5':'screen-terminal','6':'screen-shift'};
  if (map[e.key]) { nav.goto(map[e.key]); return; }
  if (e.code === 'Space' && state.currentScreen === 'screen-camera') { e.preventDefault(); camera.capture(); }
  if (state.currentScreen === 'screen-hil') {
    if (e.code === 'Enter' && !$('btn-approve').disabled) hil.approve();
    else if (e.code === 'Escape') hil.reject();
  }
}

// --- Event Delegation for Table ---
document.addEventListener('click', e => {
  const editBtn = e.target.closest('[data-edit]');
  if (editBtn) {
    const id = parseInt(editBtn.dataset.edit);
    const item = state.inventory.find(i => i.id === id);
    if (item) { state.editId = id; hil.open({item:item.item, qty:item.qty, unit:item.unit, exp_date:item.exp_date, confidence:100, ai_raw:'DIRECT HUMAN EDIT OVERRIDE.'}); }
    return;
  }
  const delBtn = e.target.closest('[data-del]');
  if (delBtn) {
    const id = parseInt(delBtn.dataset.del);
    if (confirm('Purge this record from the database?')) {
      api.del('/api/inventory/'+id+'?badge='+(state.badge||'SYSTEM')).then(() => {
        comms.add('Record purged.', 'system'); api.loadInventory();
      });
    }
  }
});

// --- Init ---
function init() {
  setInterval(tick, 1000); tick();
  api.checkHealth();
  nav.init();

  // Login
  $('btn-enter').addEventListener('click', doLogin);

  // Camera
  $('btn-capture').addEventListener('click', () => camera.capture());

  // HIL
  $('btn-approve').addEventListener('click', () => hil.approve());
  $('btn-reject').addEventListener('click', () => hil.reject());

  // Terminal
  $('btn-attach').addEventListener('click', () => $('file-input-chat').click());
  $('file-input-chat').addEventListener('change', function() { if (this.files.length) terminal.handleUpload(this.files[0]); });
  $('btn-camera-shortcut').addEventListener('click', () => nav.goto('screen-camera'));

  // Exports
  $('btn-export-csv').addEventListener('click', downloadCSV);
  if ($('btn-shift-csv')) $('btn-shift-csv').addEventListener('click', downloadCSV);
  if ($('btn-shift-audit')) $('btn-shift-audit').addEventListener('click', () => comms.add('Audit log exported.', 'system'));
  if ($('btn-close-shift')) $('btn-close-shift').addEventListener('click', () => shift.close());

  // Keys
  document.addEventListener('keydown', handleKeys);

  // Polling
  setInterval(() => { if (state.badge) api.loadInventory(); }, 5000);
}

document.addEventListener('DOMContentLoaded', init);

})();
````
